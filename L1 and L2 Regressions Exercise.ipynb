```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/erkansirin78/datasets/master/Melbourne_housing_FULL.csv')
```

```python
df.columns
```

```python
df.head()
```

```python
df.columns
```

```python
col_use = ['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount',
       'Distance','CouncilArea', 'Bedroom2','Bathroom','Car',
       'Landsize','BuildingArea','Price']
```

```python
df = df[col_use]
df
```

```python
df.isnull().sum()
```

```python
cols_to_fill_zero = ['Propertycount','Distance','Bedroom2','Bathroom','Car']
```

```python
df[cols_to_fill_zero] = df[cols_to_fill_zero].fillna(0)
```

```python
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
#Dummy Encoding

df = pd.get_dummies(df, drop_first = True)
df.head()
```

```python
X = df.drop('Price',axis = 1)
y = df['Price']
```

```python
#train_test_split

train_X,test_X,train_y,test_y = train_test_split(X,y, test_size  = 0.2, random_state = 2)
```

```python
#Linear Regression

reg = LinearRegression()

reg.fit(train_X,train_y)
```

```python
r_score_test = reg.score(test_X,test_y)
r_score_train = reg.score(train_X,train_y)
print(f'Regression Test Data Score :{r_score_test} \nLasso Reg Train Data Score :{r_score_train}')
```

```python
reg.intercept_
```

```python
reg.predict(test_X)
```

```python
#Tuning the model with Lasso Regression Method

from sklearn import linear_model 

lasso_reg = linear_model.Lasso(alpha=0.0001 , max_iter= 3000)

lasso_reg.fit(train_X,train_y)
```

```python
l_score_test = lasso_reg.score(test_X,test_y)
l_score_train = lasso_reg.score(train_X,train_y)
print(f'Lasso Reg Test Data Score :{l_score_test} \nLasso Reg Train Data Score :{l_score_train}')
```

```python
#Tuning the model with Ridge Regression Method

from sklearn.linear_model import Ridge

ridge_reg = linear_model.Ridge(alpha = 0.0001, max_iter=1000, tol=0.0001)

ridge_reg.fit(train_X,train_y)
```

```python
rg_score_test = ridge_reg.score(test_X,test_y)
rg_score_train = ridge_reg.score(train_X,train_y)
print(f'Lasso Reg Test Data Score :{rg_score_test} \nLasso Reg Train Data Score :{rg_score_train}')
```
